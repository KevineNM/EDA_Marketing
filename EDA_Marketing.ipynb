```python
###  Traitement des valeurs manquantes
df['Income'] = df['Income'].fillna(df['Income'].median())
df['Income'].head() # Permet de vérifier que la modification ne s'est appliquée qu'aux valeurs manquantes
```

```python
### Suppression des Outliers(valeurs aberrantes)
df = df[df["Year_Birth"] >= 1955]
df.describe()
```

```python
df = df[df['Income'] < 600000]
df.describe()
```

```python
### STANDARDISATION DES STATUS MATRIMAUNIAUX et EDUCATION
stand_MS = {
    'Married' : 'Married',
    'Together':'Married',
    'Single' : 'Single',
    'Widow' : 'Single',
    'Divorced':'Single',
    'Alone':'Single',
    'Absurd':'Single',
    'YOLO':'Single'
}
df['Marital_Status'] = df['Marital_Status'].map(stand_MS)

stand_Edu = {
    'Basic':'Basic',
    'Graduate':'Graduate',
    'PhD':'PostGraduate',
    'Master':'PostGraduate',
    '2nd Cycle':'UnderGraduate'
}
df['Education'] = df['Education'].map(stand_Edu)

print(f'Dataset après le nettoya de données: {df.shape}')
```

```python
### FEATURE ENGINEERING

#Age
from datetime import datetime
annee_actuelle = datetime.now().year
df['Age'] = annee_actuelle - df['Year_Birth']

#Ancienneté du client
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True, errors = 'coerce')
date_recente = df['Dt_Customer'].max()
df['Anciennete_Client'] = (date_recente - df['Dt_Customer']).dt.days

#Dépense Total du client
mnt_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Depense_Totale'] = df[mnt_cols].sum(axis=1)

#Nombre total d'achat
nbr_cols = ['NumWebPurchases', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df['Achat_Total'] = df[nbr_cols].sum(axis=1)

#Nombre total d'enfant
df['Nbr_Enfant'] = df['Kidhome'] + df['Teenhome']

#Nombre total d'acceptation aux campagnes
cmp_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
df['Nbr_Acceptation'] = df[cmp_cols].sum(axis=1)

#Revenu total par personne
df['Revenu_Par_Personne'] = df['Income']/(df['Nbr_Enfant'] + 1)

#Part d'achat en Ligne
df['Achat_Ligne'] = df['NumWebPurchases']/(df['Achat_Total'] + 0.001)
```

ANALYSE Recency Frequency Monetary (RFM)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('marketing_clean.csv')
df.head()
```

```python
df.duplicated().sum()
```

```python
#-------------------------------------------------------------------------
# CALCUL DES SCORES RFM (C'est une methode en marketing qui perment d'évaluer la valeur d'un client
#--------------------------------------------------------------------------------------------------

# 'R' pour la durée sans acheter.
df['R_score'] = pd.qcut(df['Anciennete_Client'].rank(method='first'), q=5, labels=[5, 4, 3, 2, 1], duplicates='drop')

# 'F' pour la fréquence des achats
df['F_score'] = pd.qcut(df['Achat_Total'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5])

# 'M' pour la dépense totale d'un client
df['M_score'] = pd.qcut(df['Depense_Totale'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5])

# Conversion des scores en entiers
df['R_score'] = df['R_score'].astype(int)
df['F_score'] = df['F_score'].astype(int)
df['M_score'] = df['M_score'].astype(int)

# Assemblages des scores en chaine de caractères
df['RFM_score'] =  df['R_score'].astype(str) +  df['F_score'].astype(str) +  df['M_score'].astype(str)

#Score numérique
df['RFM_score'] = (df['R_score']*0.4 + df['F_score']*0.35 + df['M_score']*0.25)


#--------------------------------------------------------------------------------------------
# SEGMENTATION RFM
#--------------------------------------------------------------------------------------------

def rfm_segment(row):
    r, f, m = row['R_score'], row['F_score'], row['M_score']
    if r >= 4 and f >= 4 and m >= 4:
        return "Champions" # Champions : récents, achètent souvent, dépensent beaucoup
    elif r >= 3 and m >= 3:
        return "Clients Loyaux" # Loyaux : fréquents et bons montants
    elif r >= 3 and f <= 2:
        return "Potentiels" # Potentiels : récents mais pas encore très fréquents
    elif r <= 2 and f >= 3:
        return "A risque"  # À risque : anciennement bons mais récence faible
    elif r <= 2 and f <= 2:
        return "Hibernants" # Hibernants : inactifs depuis longtemps
    else:
        return "En developpement"      
df['RFM_segment'] = df.apply(rfm_segment, axis=1)

#Statistique par segment
rfm_resume = df.groupby('RFM_segment').agg(
    Nbr_Client = ("ID", "count"),
    Revenu_Moyen = ("Income", "mean"),
    Depense_Moyenne = ("Depense_Totale", "mean"),
    Campagne_Accept = ("Nbr_Acceptation", "mean")
).round(2)
#print(rfm_resume.sort_values("Depense_Moyenne", ascending=True))
df.head()
```

```python
### SEGMENTATION DATA-DRIVEN AVEC KMEANS

import pandas as pd
df = pd.read_csv('marketing_rfm.csv')
df.head()
```

```python
# Identification des features
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

features = [
    "Income",
    "Depense_Totale",
    "Achat_Total",
    "Recency",
    "Age",
    "Nbr_Acceptation",
    "Nbr_Enfant",
    "Achat_Ligne"
]

X = df[features].copy()

### Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


### METHODE DU COUDE (ELBOW METHOD)

inertias = []
silhouettes = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters = k, random_state = 42, n_init = 10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    sil = silhouette_score(X_scaled, kmeans.labels_, sample_size=500, random_state=42)
    silhouettes.append(sil)
    print(f'K={k} : Inertia = {kmeans.inertia_:.0f}, Silhouette = {sil:.4f}')

### Visualisation du coude
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(list(K_range), inertias, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Nombre de cluster K')
axes[0].set_ylabel('Inertie')
axes[0].set_title('Méthode du coude - Inertie vs K')
axes[0].grid(True, alpha = 0.3)

axes[1].plot(list(K_range), silhouettes, 'gs-', linewidth=2, markersize=8)
axes[1].set_xlabel('Nombre de clusters K')
axes[1].set_ylabel('Score Silhouette')
axes[1].set_title('Score Silhouette vs K (plus haut = meilleur)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
#plt.savefig('elbow_method.png', dpi=150, bbox_inches='tight')
plt.show()
   
```

```python
# ── Application du K-Means final ──────────────────────────

K_OPTIMAL = 4

kmeans_final = KMeans(
    n_clusters=K_OPTIMAL,
    random_state=42,
    n_init=10,      # 10 initialisations différentes pour éviter
                    # de tomber sur un minimum local
    max_iter=300    # Maximum d'itérations de convergence
)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)

# ── Profil de chaque cluster ────────────────────────────────
cluster_profile = df.groupby('Cluster')[features].mean().round(2)
print('\nProfil moyen par cluster :')
print(cluster_profile)

# ── Attribution de noms métier aux clusters ────────────────
# Après avoir inspecté les profils, on nomme les clusters.
# Ces noms sont à adapter selon vos résultats réels.
# Exemple de naming basé sur les profils typiques :
cluster_names = {
    0: 'Clients Premium',      # Revenu élevé, fortes dépenses
    1: 'Clients Standard',     # Revenu moyen, comportement médian
    2: 'Clients Économes',     # Faibles dépenses, familles avec enfants
    3: 'Clients Dormants'      # Faible récence, peu d'achats
}
df['Cluster_Name'] = df['Cluster'].map(cluster_names)

#df.to_csv('marketing_clustered.csv', index=False)
print('Clustering terminé et sauvegardé.')
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('marketing_clustered.csv')

features = ['Income','Depense_Totale','Achat_Total',
            'Recency','Age','Nbr_Acceptation','Nbr_Enfant','Achat_Ligne']
X = df[features]
X_scaled = StandardScaler().fit_transform(X)

# ── PCA en 3 composantes ────────────────────────────────────
# n_components=3 : on garde 3 axes pour un scatter 3D.
# Plus on garde de composantes, plus on retient de variance.
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print(f'Variance expliquée par les 3 composantes :')
for i, v in enumerate(pca.explained_variance_ratio_):
    print(f'  PC{i+1} : {v*100:.1f}%')
print(f'  Total : {sum(pca.explained_variance_ratio_)*100:.1f}%')

# Ajout des coordonnées PCA au dataframe
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]
df['PCA3'] = X_pca[:, 2]

# ── Scatter 3D interactif ───────────────────────────────────
# px.scatter_3d : graphique 3D rotatif interactif dans le navigateur.
# Chaque point = 1 client. Couleur = cluster. Taille = Total_Spent.
fig_3d = px.scatter_3d(
    df,
    x='PCA1', y='PCA2', z='PCA3',
    color='Cluster_Name',
    size='Depense_Totale',
    hover_data=['Income', 'Age', 'Nbr_Acceptation'],
    title='Segmentation clients en 3D (PCA)',
    opacity=0.7
)
fig_3d.update_layout(height=700)
fig_3d.show()
```

```python
# ── Heatmap des profils moyens par cluster ─────────────────

cluster_profile = df.groupby('Cluster_Name')[features].mean()

# Normalisation min-max pour la heatmap uniquement (pas pour le modèle)
profile_norm = (cluster_profile - cluster_profile.min()) / (
               cluster_profile.max() - cluster_profile.min())

plt.figure(figsize=(12, 5))
sns.heatmap(
    profile_norm,
    annot=True,       # Afficher les valeurs dans chaque cellule
    fmt='.2f',        # Format : 2 décimales
    cmap='YlOrRd',    # Gradient jaune-orange-rouge
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'label': 'Score normalisé (0=min, 1=max)'}
)
plt.title('Corrélation des profils clients par segment', fontsize=14, pad=15)
plt.xticks(rotation = 45)
plt.tight_layout()
#plt.savefig('cluster_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

# ── Parallel Coordinates ───────────────────────────────────
# Le parallel coordinates plot montre TOUTES les dimensions
# simultanément. Chaque ligne = 1 client.
# Idéal pour visualiser des profils multidimensionnels.
fig_pc = px.parallel_coordinates(
    df.sample(500, random_state=42),  # Échantillon pour performance
    color='Cluster',
    dimensions=features,
    color_continuous_scale=px.colors.qualitative.Set1,
    title='Profils multidimensionnels des segments (500 clients échantillonnés)'
)
fig_pc.show()
```

```python
# ── Taux de réponse par campagne et par cluster ────────────
cmp_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3',
            'AcceptedCmp4','AcceptedCmp5','Response']

# Pour chaque campagne, on calcule le taux d'acceptation par cluster.
campaign_rates = df.groupby('Cluster_Name')[cmp_cols].mean() * 100
print('Taux de réponse aux campagnes par segment (%) :')
print(campaign_rates.round(1))

# ── Bar chart groupé ───────────────────────────────────────
campaign_melted = campaign_rates.reset_index().melt(
    id_vars='Cluster_Name',
    value_vars=cmp_cols,
    var_name='Campaign',
    value_name='Acceptance_Rate'
)

fig_camp = px.bar(
    campaign_melted,
    x='Campaign',
    y='Acceptance_Rate',
    color='Cluster_Name',
    barmode='group',  # Barres côte à côte par campagne
    title='Taux de réponse aux campagnes marketing par segment (%)',
    labels={'Acceptance_Rate': 'Taux de réponse (%)', 'Campaign': 'Campagne'}
)
fig_camp.update_layout(height=500)
fig_camp.show()

# ── Sunburst chart — vue hiérarchique ─────────────────────
# Le sunburst montre la décomposition hiérarchique :
# Segment → Education → réponse à la campagne.
df_sun = df.groupby(['Cluster_Name', 'Education']).agg(
    Count=('ID', 'count'),
    Avg_Spent=('Depense_Totale', 'mean')
).reset_index()

fig_sun = px.sunburst(
    df_sun,
    path=['Cluster_Name', 'Education'],
    values='Count',
    color='Avg_Spent',
    color_continuous_scale='RdYlGn',
    title='Répartition clients : Segment > Niveau d\'éducation'
)
fig_sun.show()

# ── Synthèse finale ────────────────────────────────────────
print('\n=== SYNTHESE DES SEGMENTS ===')
summary = df.groupby('Cluster_Name').agg(
    Nb_Clients=('ID','count'),
    Revenu_Moyen=('Income','mean'),
    Depenses_Moyennes=('Depense_Totale','mean'),
    Nb_Achats_Moyen=('Achat_Total','mean'),
    Taux_Campagnes=('Nbr_Acceptation','mean')
).round(1)
summary['Part_CA_estimee'] = (summary['Depenses_Moyennes'] * summary['Nb_Clients'])
summary['Part_CA_pct'] = (summary['Part_CA_estimee'] / summary['Part_CA_estimee'].sum() * 100).round(1)
print(summary.sort_values('Depenses_Moyennes', ascending=False))
```
